# 2. Data Cleaning Prep


```python
ax = plt.axes()
sns.heatmap(books.isnull().transpose(), annot=True, cbar=False, cmap="coolwarm")
ax.set_title("Patterns of missing values in the dataset")
```

```python
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
                      cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
book_missing = books.dropna(subset=['description', 'num_pages', 'average_rating', 'published_year'])
```

```python
book_missingg = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]
```

```python
# Create the value counts DataFrame
value_counts_df = book_missing["age_of_book"].value_counts().reset_index()
value_counts_df.columns = ['Age_of_Book', 'Count']
value_counts_df = value_counts_df.sort_values(by='Age_of_Book')

# Create a cleaner visualization
plt.figure(figsize=(14, 8))
ax = sns.barplot(x='Age_of_Book', y='Count', data=value_counts_df, color='steelblue')

# Improve the x-axis by showing fewer labels
plt.xticks(rotation=45)
# Only show every nth label to reduce clutter
n = 5  # Adjust this value as needed
for i, label in enumerate(ax.xaxis.get_ticklabels()):
    if i % n != 0:
        label.set_visible(False)

plt.title('Frequency of Books by Age', fontsize=16)
plt.xlabel('Age of Book (years)', fontsize=12)
plt.ylabel('Number of Books', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
Top_popular_categories = book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
N = 15
top_n_categories = Top_popular_categories.head(N)
plt.figure(figsize=(14, 10))
colors = sns.color_palette("YlOrRd_r", N)
bars = plt.barh(top_n_categories['categories'], top_n_categories['count'], color=colors)

# Add count labels
for bar in bars:
    width = bar.get_width()
    plt.text(width + 5, bar.get_y() + bar.get_height()/2, f'{int(width)}', 
             ha='left', va='center', fontsize=10)

# Add labels and title
plt.title('Distribution of Books Across Top {} Categories'.format(N), fontsize=16)
plt.xlabel('Number of Books', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
book_missing.loc[:, "words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
# Set figure size
plt.figure(figsize=(8, 8))

# Create scatterplot with hue based on missing descriptions and size based on number of pages
sns.relplot(
    data=book_missing,
    x="words_in_description", 
    y=book_missing.index, 
    color="royalblue", 
    hue="missing_description",   # Coloring based on missing description status
    size="num_pages",            # Size of points based on number of pages
    sizes=(10, 200),             # Size range for the points
    height=10,                   # Height of the plot
    aspect=1.2                   # Aspect ratio (width:height ratio)
)

# Add title and labels
plt.title("Distribution of Word Counts in Book Descriptions", fontsize=14)
plt.xlabel("Number of Words in Description", fontsize=12)
plt.ylabel("Book Index", fontsize=12)

# Show the plot
plt.show()
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
book_missing_25_words
```

```python
(
    book_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("../data/processed/books_cleaned.csv", index=False)
)
```
